```python
#VISUALIZATION
#CHECK HOW SEPALLENGTHCM AFFECTS THE SPECIES USING CROSSTABS
pd.crosstab(iris["SepalLengthCm"],iris["Species"],margins=True)
```

```python
#SCATTER PLOTS FOR Species VS SepalLengthCm
plt.scatter(x=iris["Species"],y=iris["SepalLengthCm"])
plt.show()            
```

```python
# Pairplot with species hue
sn.pairplot(iris, hue='Species', palette='rainbow')
plt.show()
```

```python
correlation_matrix = iris.iloc[:, :-1].corr()
```

```python
plt.figure(figsize=(8, 6))
sn.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```
